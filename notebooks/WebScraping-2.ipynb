# Scrape the tables from the URLs

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
headers = {
    "User-Agent": "MyResearchBot/1.0 (contact: adhainje1987@gmail.com)"
}
page = requests.get(url, headers=headers)
# page = requests.get(url)
content = BeautifulSoup(page.text, "html")
```

```python
table = content.find("table", class_="wikitable sortable")
titles = [t.text.strip() for t in table.find_all("th")]
df = pd.DataFrame(columns = titles)
df
```

```python
column_data = table.find_all("tr")
for row in column_data:
    row = row.find_all("td")
    individual_column_data = [t.text.strip() for t in row]
    if(len(individual_column_data) > 0):
        length = len(df)
        df.loc[length] = individual_column_data
df.head()
```
